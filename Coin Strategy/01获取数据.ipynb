```python
import ccxt
import json
with open('config.json','r') as f:
    config = json.load(f)
exchange = ccxt.binance({
    'options':{
        'adjustForTimeDifference':True,
        'defaultType':'future',
    },
    'apiKey':config['apiKey'],
    'secret':config['secret'],
})
print(exchange.fetch_balance())
```

```python
import pandas as pd
symbol = 'BTCUSDT'
period = '3m'
ohlcv = exchange.fetch_ohlcv(symbol, period)
klines = pd.DataFrame(ohlcv, columns=['Date', 'Open', 'High', 'Low', 'Close', 'Volume'])
klines['time'] = pd.to_datetime(klines['Date'], unit='ms')
klines['time'] = klines['time'].dt.tz_localize('UTC').dt.tz_convert('Asia/Shanghai')
klines['time'] = klines['time'].dt.tz_localize(None)
klines.set_index('time', inplace=True)
klines.insert(0, 'Symbol', symbol)
# CSV
def export_to_csv(klines, symbol, period):
    """
    CSVsymbol
    :param klines: 
    :param symbol: 'BTC/USDT'
    :param period: '2h'
    """
    # BTCUSDT+2h.csv
    filename = f"{symbol.replace('/', '')}+{period}.csv"
    klines.to_csv(filename)
    print(f"Data exported to {filename}")
# 
export_to_csv(klines, symbol, period)
klines
```
